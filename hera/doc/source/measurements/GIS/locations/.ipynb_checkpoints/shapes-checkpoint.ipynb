# Shapes

This tools helps managing manipulations on polygons and points that may be saved at the database. It may be usefull to save coordinates of known locations such as cities or points of interest to the database. These areas of interest may be used for queries, for example.

## Datalayer

In order to add a new shape to the database, use the addShape function. The function must receive an unused name for the new location, and the location itself.

A point can be induced either as a shapely.geometry.Point or a list.

```python
from shapely import geometry
from hera.measurements.GIS.locations.shapes import datalayer as shapeDatalayer

shape = shapeDatalayer(projectName="Documentation", databaseNameList=None, useAll=False,publicProjectName="Shapes")
Haifa_Port = geometry.Point([200000,748000])
# Haifa_Port = [200000,748000] this syntax would work as well.

shape.addShape(name="Haifa_Port", Shape=Haifa_Port)
```

A polygon can be induced either as a shapely.geometry.Polygon, a list of points or a list of x coordinated and y coordinates.

```python
Bahai_Gardens = geometry.Polygon([[199220,747015],[199333,746947],[198957,746277],[198779,746438]])
# Bahai_Gardens = [[199220,747015],[199333,746947],[198957,746277],[198779,746438]] this syntax would work as well.
# Bahai_Gardens = [[199220,199333,198957,198779],[747015,746947,746277,746438]] this syntax would work as well.

shape.addShape(name="Bahai_Gardens", Shape=Bahai_Gardens)
```

The shapes may be loaded using getShape function by addressing their names.

```python
type(shape.getShape(name="Haifa_Port"))
```

The shape’s coordinates and type of shapely geometry may also be loaded.

```python
points, geometry_type = shape.getShapePoints("Bahai_Gardens")
```

## Presentation layer

Saved shapes may be easily plotted using the tool's presentation layer.

The default color for the locations is black, and the default marker for points is a star. They can be changed by setting the parameters “color” and “marker”, the same as in mayplotlib.pyplot.plot.

```python
import matplotlib.pyplot as plt
import geopandas
from hera.measurements.GIS.locations.topography import datalayer as topoDatalayer

topography = topoDatalayer(projectName="Documentation", FilesDirectory="/home/ofir/Development/pyhera/hera/doc/source/measurments/GIS/locations/examples", 
                           databaseNameList=None, useAll=False,publicProjectName="Topography",Source="BNTL")
#example_data = topography.getDocuments(Region="Haifa")[0].getData()
example_data = geopandas.read_file("examples/Example.shp")

fig, ax = plt.subplots(1,1)
ax = example_data.plot()
shape.presentation.plot(names=["Haifa_Port", "Bahai_Gardens"], ax=ax) 
plt.ylim(745000,750000)
plt.xlim(195000,203000)
```
