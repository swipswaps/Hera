# Data Manipulations

Here we will present some built in manipulations one can perform on geopandas DataFrames.

```python
from hera import GIS
from shapely.geometry import Polygon
import matplotlib.pyplot as plt

manipulator = GIS.dataManipulations()
datalayer = GIS.GIS_datalayer(projectName = "Examples", FilesDirectory = "/home/ofir/tmp/BNTL-HAIFA")
```

## Creating a Rectangle

In many cases we would like to create a polygon in the shape of a rectangle.
There is a function that saves the effort of writing
by building it using the bottom left and upper right coordinates only:

```python
PartOfHaifa = manipulator.makePolygonFromEndPoints(points=[195000,744000,200000,749000])
```

## Intersection with Polygons

In order to decrease the spacial boundaries of a dataframe to the boundaries of a polygon,
a built in function may be used.
For example, a part of Haifa was saved as a polygon in the dataframe.
The polygon and a dataframe that contains it can be loaded using the GIS datalayer.

```python
#datalayer.addGeometry(Geometry=PartOfHaifa, name="PartOfHaifa")
dataframe = datalayer.getGISDocuments(Geometry="PartOfHaifa")[0][0].getData()

x,y = PartOfHaifa.exterior.xy

fig, ax = plt.subplots()

ax = dataframe.plot()
ax.plot(x,y, color="black")

plt.show()
```

These shapely polygon and geopandas dataframe can be given as parameters to a function that returns a new dataframe, with the boundaries of the polygon.  

```python
new_dataframe = manipulator.PolygonDataFrameIntersection(dataframe=dataframe, polygon=PartOfHaifa)

new_dataframe.plot()
plt.show()
```

## Convex Polygons

geopandas has a function called convex_hull, which return a polygon that includes all the polygons of a dataframe.
However, sometimes the dataframe holds several remote areas of adjust polygons,
and we would like to get the convex polygon of each area.
For those case, we can use the following function.

We will use it, as an example, on a dataframe of buildings of Katsrin.

```python
GISdatalayer = GIS.GIS_datalayer(projectName="Examples", FilesDirectory="/home/ofir/Projects/2020/GIS")
buildings = GISdatalayer.getGISDocuments(points=[263000,765000,266000,768000],CutName="KatsrinBuildings",mode="Buildings")[0].getData() 
polygons = manipulator.ConvexPolygons(buildings)
polygons.plot()
polygons
```

We may choose different buffers to decide which distances between areas leads to seperations of polygons. The defaut buffer is 100 m.

```python
largeBuffer = manipulator.ConvexPolygons(buildings, buffer=1000)
smallBuffer = manipulator.ConvexPolygons(buildings, buffer=10)
largeBuffer.plot()
smallBuffer.plot()
```
