Here we will present some built in functions that convert the data to other formats.

```python
from hera import GIS
from shapely.geometry import Polygon
import matplotlib.pyplot as plt

converter = GIS.utils.STL.convert(projectName = "Example", FilesDirectory = "/home/ofir/tmp/BNTL-HAIFA")
```

## Convertion to stl

The data can be converted from a geodataframe format to an stl format using a buit in convertion function.
The data can be induced as a geodataframe or as a name of a saved polygon in the database.
If the data is a dataframe, the whole dataframe is converted to stl.
If it is a polygon, a dataframe that contains only the requested polygon is being made and then converted to stl.

The stl string is saved by default as a file and the path to the file is added to the database.
The path is by default the FilesDirectory, but other paths may be given.
Any parameters and corresponding values may be given to the function too, and they are induced to the new metadata document.

A name for the new file must be given even if the file is not saved, because it is used in the stl string.

The size of cells in the dataframe used for the convention is 50X50 meters by default. 
It can be changed by inducing another value as the parameter dxdy.

In the next example the file is not saved. We would create an stl string for a polygon that holds part of Haifa.

```python
polyon = "PartOfHaifa"
NewFileName = "PartOfHaifaSTL"

stlstring, data = converter.toSTL(data=polyon, NewFileName=NewFileName, save=False, dxdy=100)
```

The function returns a the stl string and a dataframe that holds the x and y coordinates and their hieghts.
It also holds the max values of the x, y and z axis.
