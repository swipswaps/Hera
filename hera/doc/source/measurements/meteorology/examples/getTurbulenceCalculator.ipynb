# Getting turbulence calculator
This notebook shows how to get turbulence calculator with pera

```python
from pyhera.hera import meteo
```

After importing meteorological, you can get the turbulence caclculator that fits your requirments.
Below we see an example of getting the turbulence calculator of data between 2 dates of Haifa campaign in station:'Check_Post', instrument:'Sonic', height:9(m).

```python
projectName = 'Haifa'
station = 'Check_Post'
instrument = 'Sonic'
height = 9
start = '2015-08-01' # can be as pandas Timestamp: start = pandas.Timestamp('2015-08-01')
end = '2015-08-02' # can be  as pandas Timestamp: end = pandas.Timestamp('2015-08-02')
samplingWindow = '5min'
trbCalc = meteo.getTurbulenceCalculator(projectName=projectName,
                                        samplingWindow=samplingWindow,
                                        station=station,
                                        instrument=instrument,
                                        height=height,
                                        start=start,
                                        end=end
                                        )
```

After getting the turbulence calculator you can start calculating the required variables.
Below we see an example of calculating the zOverL, and the wind speed.
(Using the compute mode of getting the data not from the database and not saving the result to the database)

```python
data = trbCalc.zOverL().wind_speed().compute(mode='not_from_db_and_not_save')
print(data)
```
