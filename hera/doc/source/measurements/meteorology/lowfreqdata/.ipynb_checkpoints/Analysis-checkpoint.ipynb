## Analysis 

Sometimes we need to perform simple queries on the date. 
The addDatesColumns function in the lowfreqdata.analysis module  adds date and time columns to the dataframe from the date index column. These fields can,
for example, when you want to extract data by different time slices (by season, year, hour, etc.)

First, we get the data

```python
stationName = 'BET DAGAN'
from hera.measurements.meteorology import lowfreqdata 
datadb = lowfreqdata.lowfreqDataLayer.getStationDataFromDB(StationName=stationName)
```

```python
datadb.getData().compute().head()
```

Now we import the analysis function 

```python
lowfreqdata.analysis.addDatesColumns(datadb.getData()).compute().head()
```
